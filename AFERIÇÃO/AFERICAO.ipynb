```python
#a)
def contaEspacos(s):
    if s == '':
        c = 0
    elif s[0] == ' ':
        c = 1 + contaEspacos(s[1:])
    else:
        c = contaEspacos(s[1:])
    return c

#b)
def ordenada(l):
    if len(l) <= 1:
        b = True
    elif l[0] <= l[1]:
        b = ordenada(l[1:])
    else:
        b = False
    return b

#c)
def existeElem(l,e):
    if l == []:
        b = False
    elif l[0] == e:
        b = True
    else:
        b = existeElem(l[1:],e)
    return b
```

```python
#a)
def maiorDivisor(n):
    res = {}
    for i in range(1,n+1):
        for d in range(1,10):
            if i%d == 0:
                res[f'{i}'] = d
    return res

#b)
def mmc(a,b):
    c = a
    d = b
    m = 2
    n = 2
    while c != d:
        if c > d:
            d = b*n
            n += 1
        elif d > c:
            c = a*m
            m += 1
    return c

#c)
def f2c(reg):
    res = {}
    for f in reg:
        tempf = reg[f]
        tempc = (tempf - 32)/1.8
        res[f] = tempc
    return res

#d)
def reduxInt(n):
    while len(str(n)) > 1:
        s = str(n)
        n = 0
        for a in s:
            n += int(a)
    return n
```

```python
#a)
def listar_unidades(hosp):
    res = []
    for unid in hosp:
        res.append(unid[0])
    return(res)

#b)
def camas_disponiveis(hosp):
    n = 0
    for unid in hosp:
        for cama in unid[2]:
            if not cama[1]:
                n += 1
    return n

#c)
def listar_medicos_unidade(hosp,nomeu):
    res = []
    for unid in hosp:
        if unid[0] == nomeu:
            for med in unid[1]:
                res.append(med)
    return res

#d)
def existe_medico(hosp,nomeu,nomem):
    b = False
    for unid in hosp:
        if unid[0] == nomeu:
            for med in unid[1]:
                if med[1] == nomem:
                    b = True
    return b

#e)
def adicionar_medico(hosp,nomeu,nomem):
    for unid in hosp:
        if unid[0] == nomeu:
            nomesm = [x[1] for x in unid[1]]
            ids = [x[0] for x in unid[1]] 
            if nomem not in nomesm:
                if ids == []:
                    idn = 1
                else:
                    idn = max(ids) + 1
                unid[1].append((idn,nomem))
    return hosp

#f)
def ocupar_cama(hosp,nomeu,idcama):
    for unid in hosp:
        if unid[0] == nomeu:
            for i in range(len(unid[2])):
                cama = unid[2][i]
                if cama[0] == idcama and not cama[1]:
                    unid[2][i] = (cama[0], True)
    return hosp

#g)
def guardar_hospital(hosp,nomef):
    import json
    f = open(f'./{nomef}.json','w')
    json.dump(hosp,f)
    f.close()

#h)
def carregar_hospital(nomef):
    import json
    f = open(f'./{nomef}.json','r')
    hosp = json.load(f)
    f.close()
    return hosp

#i)
def camas_ocupadas(hosp):
    n = 0
    for unid in hosp:
        n += len(unid[2])
    ocupa = n - camas_disponiveis(hosp)
    return (ocupa/n)*100

#j)
def unidades_ocupacao_critica(hosp):
    res = []
    for unid in hosp:
        n = 0
        for cama in unid[2]:
            if cama[1] == True:
                n += 1
        ocupa = n/len(unid[2])
        if ocupa >= 0.8:
            res.append((unid[0],ocupa*100))
    return res
```
